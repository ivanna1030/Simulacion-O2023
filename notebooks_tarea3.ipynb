**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np

sp.var("x")


def f(x) :
    return 1260 + 117*x - 9*x**2
df = sp.diff(f(x), x)
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace(0, 13, 1000)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.plot(xn, f(xn))
for c in pc :
    plt.plot(c, f(c), "*r", ms=10)
```

La función tiene 1 punto crítico en 6.5 y es un máximo global.

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
import sympy as sp
import numpy as np

sp.var("x")


def f(x) :
    return (1/4*x**4) + (1/3*x**3) - x**2
df = sp.diff(f(x), x)
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace(-3, 1.7, 1000)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.plot(xn, f(xn))
for c in pc :
    plt.plot(c, f(c), "*r", ms=10)
```

La función tiene 3 punto críticos en -2, 0 y 1.
- El punto crítico en -2 es un mínimo global.
- El punto crítico en 0 es una máximo local.
- El punto crítico en 1 es un mínimo local.

El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

Al ser triángulo isósceles, dos lados son iguales y uno es diferente, por lo que se genera la siguiente ecuación, en la que se igualan a su perímetro:
$$ x + 2y = 10 $$

Para sacar "r", podemos despejar x, ya que ésta es dos veces el radio:
$$ x = 10 - 2y $$
$$ r = \frac{x}{2} = \frac{10-2y}{2} = 5-y $$

Podemos utilizar el teorema de pitágoras para sacar "h", ya que podemos usar "y" como hipotenusa y "r" como un cateto:
$$ y^2 = h^2 + r^2 $$
$$ h^2 = y^2 - r^2 $$
$$ h^2 = y^2 - (5-y)^2 $$
$$ h^2 = y^2 - (25 - 10y + y^2) $$
$$ h^2 = y^2 - 25 + 10y - y^2 $$
$$ h^2 = -25 + 10y $$
$$ h = \sqrt{10y - 25} $$

Podemos sustituir cada una de las variables en la fórmula del cono:
$$ V = \frac{1}{3}\pi r^2 h = \frac{1}{3}\pi (5-y)^2 \sqrt{10y - 25} $$

Ahora podemos encontrar el mayor volumen posible, respecto a "y":

```python
import math as mt

sp.var("y")

def f(y) :
    return (1/3) * mt.pi * ((5-y)**2) * ((10*y - 25)**(1/2))

df = sp.diff(f(y), y)
pc = sp.solve(df, y)
pc
```

```python
yn = np.linspace(2.6, 5.4, 1000)

plt.figure()
plt.plot(yn, f(yn))
for c in pc :
    plt.plot(c, f(c), "*r", ms=10)
```

Aunque la función tiene 2 puntos críticos, en 3 y en 5, solamente utilizaremos el del 3, ya que es un máximo y nosotros buscamos maximizar el volumen del cono.
Sustituimos el valor de "y" que es 3, en la ecuación:
$$ x = 10 - 2y $$
$$ x = 10 - 2(3) $$
$$ x = 10 - 6 $$
$$ x = 4 $$

Podemos concluir que el triángulo tiene un perímetro de 10 cm y sus lados miden 3 cm, 3 cm y 4 cm.

Para finalizar, sustituimos el valor de "y" que es 3 en la fórmula del volumen del cono:
$$ V = \frac{1}{3}\pi r^2 h = \frac{1}{3}\pi (5-y)^2 \sqrt{10y - 25} $$
$$ V = \frac{1}{3}\pi r^2 h = \frac{1}{3}\pi (5-3)^2 \sqrt{10(3) - 25} $$

Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

Creamos la ecuación con los dos postes ("x"), el travesaño ("y") y el perímetro de la portería (6 metros):
$$ 2x + y = 6 $$

La fórmula para encontrar el área maxima es:
$$ A = xy $$

Despejamos "y" de la ecuación:
$$ y = 6 - 2x $$

Y la sustituimos en la fórmula del área:
$$ A = x(6 - 2x) $$
$$ A = 6x - 2x^2 $$

Ahora podemos encontrar el área máxima:

```python
def f(x) :
    return 6*x - 2*x**2

df = sp.diff(f(x), x)
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace(0, 3, 1000)

plt.figure()
plt.plot(xn, f(xn))
for c in pc :
    plt.plot(c, f(c), "*r", ms=10)
```

La función tiene un punto crítico en 1.5, el cuál es un máximo, por lo que podemos calcular el valor de "y" sustituyendo "x" con el valor de 1.5 en la ecuación:
$$ y = 6 - 2x $$
$$ y = 6 - 2(1.5) $$
$$ y = 6 - 3 $$
$$ y = 3 $$

Para finalizar, sustituimos los valores en la fórmula del área para encontrar el área máxima:
$$ A = xy $$
$$ A = (1.5)(3) $$
$$ A = 4.5 $$
